## Imputation

### **Date standardization** & **Validation**

### Step 1: Convert 'Purchase Date' to datetime format

```python
# We convert different date formats to a consistent pandas datetime format. Invalid dates become NaT.

df['Purchase Date'] = pd.to_datetime(df['Purchase Date'], errors='coerce')
print(df['Purchase Date'])
```

### **Backward/Forward Fill**

```python
# df['Purchase Date'] = df['Purchase Date'].fillna(method='ffill')  # forward fill (Deprecated)
# df['Purchase Date'] = df['Purchase Date'].fillna(method='bfill')  # backward fill (Deprecated)
# df['Purchase Date'].ffill()
df['Purchase Date'].bfill()
```

### Step 2: Fix invalid emails

```python
import re # RegEx module (Pythons built-in regular expression library).
import numpy as np
def fix_email(email):
    if isinstance(email, str):
        pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
        if re.match(pattern, email):
            return email
    return np.nan

df['Email'] = df['Email'].apply(fix_email)
print(df['Email'])
```

### **Data Handling** 

### Step 3: Fill missing 'Quantity' and 'Price' with median values

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price'] = df['Price'].fillna(df['Price'].mean())

print(df[['Quantity', 'Price']])
```

### Step 4: Remove exact duplicates

```python
# Drops rows that are exact duplicates of others.
df = df.drop_duplicates()
print(df.shape)
```

### Step 5: Fix negative quantities

```python
df.loc[df['Quantity'] < 0, 'Quantity'] = df['Quantity'].median()

print(df.loc[df['Quantity'] < 0])  # Should be empty now
```

### Step 6: Fill missing 'Product' names with 'Unknown'

```python
df.loc[:, 'Product'] = df['Product'].fillna('Unknown')
print(df['Product'])
```

### Step 7: Replace zero prices with median price

```python
df.loc[df['Price'] == 0, 'Price'] = df['Price'].median()

print(df.loc[df['Price'] == 0])  # Should be empty now
df['Price']
```

### **Cleaned Dataset**
### Step 8: Reset index

```python
df = df.reset_index(drop=True)
print("Final clean data sample:")
df.head(20)
```
